# Feature Scaling (Preprocessing)

```python
from sklearn.preprocessing import StandardScaler
```

```python
print("mean of X",X.values.mean())
print("Std of X",X.values.std())
```

```python
XA=np.array(X)
YA=np.array(Y)
```

```python
scaler=StandardScaler()
XA=scaler.fit(XA).transform(XA)    
#if you use transform than output will be array otherwise it is StandardScaler
```

```python
print("mean of XA",round(XA.mean()))
print("Std of XA",XA.std())
```

```python
type(XA)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#if random_state fix than data is not shuffled it is changes it is checked by removing and print Xtrain
xtrain,xtest,ytrain,ytest=train_test_split(XA,YA,test_size=.30,random_state=101)
xtrain
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
#if you use 500 than after running learner you get warning
#13 it is not the part of input
trainer=MLPClassifier(hidden_layer_sizes=(13,13,13),max_iter=500,activation="logistic") 
```

```python
learner=trainer.fit(xtrain,ytrain)  #see warning here
```

```python
#here we use 1000 iteration so optimization converged
#13 it is not the part of input
trainer=MLPClassifier(hidden_layer_sizes=(13,13,13),max_iter=1000,activation="logistic") 
```

```python
learner=trainer.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score,jaccard_similarity_score
```

```python
Yp=learner.predict(xtest)
Ya=ytest
```

```python
acc=accuracy_score(Yp,Ya)*100
jss=jaccard_similarity_score(Yp,Ya)*100
print("jss {} acc {}".format(jss,acc))
```
