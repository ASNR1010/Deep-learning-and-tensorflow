```python
X_input=mydata["Message"]    #selecting feaures or inputs
Y_out=mydata["Target"]       #selscting labels or outputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
#train_test_split is used to convert data explicitly
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X_input,Y_out,test_size=.30,random_state=101)
```

```python
#analysis of shapes of training and testing
print("Shape of Xtrain is {}".format(Xtrain.shape))
print("Shape of Ytrain is {}".format(Ytrain.shape))
print("Shape of Xtest is {}".format(Xtest.shape))
print("Shape of Ytest is {}".format(Ytest.shape))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cont_vect=CountVectorizer()
```

```python
training_data=cont_vect.fit(Xtrain).transform(Xtrain)
testing_data=cont_vect.transform(Xtest)
```

```python
print(f"Shape of training_data is {training_data.shape}")
```
