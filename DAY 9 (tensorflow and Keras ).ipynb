```python
import tensorflow as tf
import pandas as pd
from tensorflow.keras import layers
from tensorflow import keras
```

```python
mydata=pd.read_csv(r"C:\Users\ANKIT SINGH\Desktop\internship\DataSets-master\500_Person_Gender_Height_Weight_Index.csv")
```

```python
mydata[:2]
```

```python
X=mydata.iloc[:,0:3]
Y=mydata.iloc[:,3]   #output is index
```

```python
X      
```

```python
Y  #index values separate
```

```python
Index_name=pd.Series(["Extremely Weak","Weak","Normal","Overweight","Obesity","Extreme Obesity"])
Index_name
```

```python
X["Gender"]=X["Gender"].map({"Male":0,"Female":1})
```

```python
XA=X.values
YA=Y.values
```

```python
XA
```

```python
YA
```

```python
print("Type od XA {} and Type of YA {}".format(type(XA),type(YA)))
```

```python
Yc=tf.keras.utils.to_categorical(YA,num_classes=6,dtype="float32")
Yc
```

```python
type(Yc)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(XA,Yc,test_size=.30,random_state=101)
```

```python
model=tf.keras.Sequential()
```

```python
model.add(layers.Dense(20,input_dim=3,activation="relu"))
model.add(layers.Dense(20,activation="relu"))
model.add(layers.Dense(6,activation='softmax'))

#compiling model
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=500,batch_size=100)
```

```python
score=model.evaluate(xtest,ytest)
```
